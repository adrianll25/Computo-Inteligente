```python
#Remplaza los datos o celdas vacias con la media
#Media promedio de una serie de valores

m = df_csv["age"].mean()
df_csv["age"].fillna(m, inplace = True)
```

```python
#Media: valor medio de una serie de valores ordenados

m = df_csv["age"].median()
df_csv["age"].fillna(m, inplace = True)
```

```python
df_csv
```

```python
df_csv = copia.copy()
df_csv
```

```python
#Moda: Valor mas popular de una serie de valores
m = df_csv["age"].mode()[0]
print(m)
df_csv["age"].fillna(m, inplace = True)
df_csv
```

```python
df_csv
```

```python
#Datos equivocados
#titanic: clases: 1, 2, 3 <- valores permitidos. Valor equivocados 0, 4, 5, etc.
#Para modificar un registro: registro 1
#df.loc[1, "pclass"] = 3
#Para modificar toda la columna

for i in df_csv.index:
    if df_csv.loc[i, "pclass"] > 3 or df_csv.loc[i, "pclass"] < 1:
        if df_csv.loc[i, "class"] == "Third":
            df_csv.loc[i, "pclass"] = 3
        elif df_csv.loc[i, "class"] == "Second":
             df_csv.loc[i, "pclass"] = 2
        elif df_csv.loc[i, "class"] == "First":
             df_csv.loc[i, "pclass"] = 1
```

```python
df_csv
```

```python
import pandas as pd
```

```python
#Para borrar datos equivocados y eliminar 
for i in df_csv.index:
    if df_csv.loc[i, "age"] < 5.0:
        print(i)
        df_csv.drop(i, inplace = True)
```

```python
df_csv
```

```python
#Para eliminar renglones duplicados
print(df_csv.duplicated())
```

```python
#Para eliminar duplicados
df_csv.drop_duplicates(inplace = True)
```
